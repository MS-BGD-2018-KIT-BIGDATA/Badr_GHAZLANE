```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import itertools 
from multiprocessing import Pool
import functools
```

```python
def getSoupFromURL(url, method='get', data={}):

  if method == 'get':
    res = requests.get(url)
  elif method == 'post':
    res = requests.post(url, data=data)
  else:
    return None

  if res.status_code == 200:
    soup = BeautifulSoup(res.text, 'html.parser')
    return soup
  else:
    return None
```

```python

def get_distance(city1,city2):
    
    url_distance = 'https://maps.googleapis.com/maps/api/distancematrix/json?units=imperial&origins=' + city1 + '&destinations='+ city2+'&key=AIzaSyBKp5mwxpY27I0DOWmDn8EfjaUklYZl1KI'
    r = requests.get(url_distance)
    return r.json()['destination_addresses'],r.json()['rows'][0]['elements'][0]['distance']['value']
```

```python
def get_list_top_cities():
    url_cities = 'https://lespoir.jimdo.com/2015/03/05/classement-des-plus-grandes-villes-de-france-source-insee/'
    soup = getSoupFromURL(url_cities)
    
    tab_cities = []
    for i in range(1,20):
        tab_cities.append(soup.find_all('tr')[i].find_all('td')[1].text.replace('\n            ',''))
    return tab_cities
    
```

```python
a = get_list_top_cities()
```

```python
d = pd.DataFrame(0, index=a, columns=a)
for city in a:
    for citybis in a: 
        d[city].loc[citybis] = get_distance(city,citybis)[1]
```
