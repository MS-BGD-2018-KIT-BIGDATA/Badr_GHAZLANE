```python
def getMedId():
    url = 'https://www.open-medicaments.fr/api/v1/medicaments?query=ibuprof%C3%A8ne'
    r = requests.get(url)
    r.json()
    
    if r.status_code == 200:
        med_id = pd.Series([med['codeCIS'] for med in r.json()])
        return med_id
    else:
        return 'Authentification problem or med doesn\'t not exist'
```

```python
def construct_url():
    list_med_id = getMedId()
    list_med_id = list_med_id.apply(lambda x: 'https://www.open-medicaments.fr/api/v1/medicaments/' + str(x) ) 
    
    return list_med_id
```
